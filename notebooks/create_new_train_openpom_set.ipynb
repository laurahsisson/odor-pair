```python
import pandas as pd

df_act = pd.read_csv("./curated_GS_LF_merged_4983.csv")
```

```python
df_test = pd.read_csv("./cleaned_non_stereo_test.csv")
```

```python
df_act.head()
```

```python
df_test.head()
```

```python
test_smiles_set = set()

def get_test_smiles(smiles):
    list_compo = smiles.split(".")
    test_smiles_set.update(list_compo)
```

```python
df_test['combined_smiles_non_stereo'].apply(lambda x: get_test_smiles(x))
```

```python
len(test_smiles_set)
```

```python
df_new_open_pom_train = df_act[~df_act['nonStereoSMILES'].isin(test_smiles_set)]
```

```python
df_new_open_pom_train.head()
```

```python
df_new_open_pom_train_ = df_new_open_pom_train.drop(columns=["nonStereoSMILES","descriptors"])
odors_df = pd.DataFrame(df_new_open_pom_train_.sum().sort_values(ascending=False), columns=['sum'])
odors_df.query('sum>=0')
```

```python
list(df_new_open_pom_train.columns)
```

```python
# openpom uses mint instead of minty as label
df_new_open_pom_train = df_new_open_pom_train.rename(columns={"mint":"minty"})
```

```python
df_new_open_pom_train.to_csv("new_open_pom_train.csv", index=False)
```
